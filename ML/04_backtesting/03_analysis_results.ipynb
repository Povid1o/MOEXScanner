```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
```

## –ó–∞–≥—Ä—É–∑–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤


```python
BACKTEST_DIR = Path('data') / 'backtest'
ticker = 'SBER'

portfolio = pd.read_parquet(BACKTEST_DIR / f"{ticker}_backtest_portfolio.parquet")
metrics = pd.read_csv(BACKTEST_DIR / f"{ticker}_backtest_metrics.csv")

print(f"‚úÖ –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã: {len(portfolio)} –∑–∞–ø–∏—Å–µ–π")
print(f"\n–ú–µ—Ç—Ä–∏–∫–∏:")
print(metrics.T)
```

## –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è Equity Curve


```python
fig, ax = plt.subplots(figsize=(14, 6))

ax.plot(portfolio['equity'], label='Strategy', linewidth=2, color='blue')
ax.plot(portfolio['buy_hold_equity'], label='Buy & Hold', linewidth=2, color='orange', alpha=0.7)

ax.set_title(f'Equity Curve - {ticker}', fontsize=14, fontweight='bold')
ax.set_ylabel('Portfolio Value', fontsize=12)
ax.set_xlabel('Time', fontsize=12)
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("‚úÖ Equity curve –ø–æ—Å—Ç—Ä–æ–µ–Ω–∞")
```

## –ê–Ω–∞–ª–∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π


```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏
axes[0].hist(portfolio['strategy_returns'].dropna(), bins=50, alpha=0.7, color='blue', edgecolor='black')
axes[0].axvline(x=0, color='red', linestyle='--', linewidth=2)
axes[0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ Strategy Returns', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Returns', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].grid(True, alpha=0.3)

# –õ—É—á—à–∏–µ –∏ —Ö—É–¥—à–∏–µ –¥–Ω–∏
best_days = portfolio['strategy_returns'].nlargest(10)
worst_days = portfolio['strategy_returns'].nsmallest(10)

axes[1].barh(range(10), best_days.values, color='green', alpha=0.7, label='Top 10 Best')
axes[1].barh(range(10, 20), worst_days.values, color='red', alpha=0.7, label='Top 10 Worst')
axes[1].set_title('–¢–æ–ø 10 –õ—É—á—à–∏—Ö/–•—É–¥—à–∏—Ö –î–Ω–µ–π', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Returns', fontsize=11)
axes[1].legend()
axes[1].grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()

print("‚úÖ –ê–Ω–∞–ª–∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≥–æ—Ç–æ–≤")
```

## –ò—Ç–æ–≥–æ–≤–∞—è —Å–≤–æ–¥–∫–∞


```python
print("\n" + "="*60)
print(f"üìä –ò–¢–û–ì–û–í–ê–Ø –°–í–û–î–ö–ê –ë–≠–ö–¢–ï–°–¢–ê - {ticker}")
print("="*60)
print(f"–ù–∞—á–∞–ª—å–Ω—ã–π –∫–∞–ø–∏—Ç–∞–ª: {portfolio['equity'].iloc[0]:,.2f}")
print(f"–ö–æ–Ω–µ—á–Ω—ã–π –∫–∞–ø–∏—Ç–∞–ª: {portfolio['equity'].iloc[-1]:,.2f}")
print(f"–î–æ—Ö–æ–¥–Ω–æ—Å—Ç—å: {metrics['total_return'].iloc[0]:.2%}")
print(f"Buy & Hold: {metrics['buy_hold_return'].iloc[0]:.2%}")
print(f"Alpha: {(metrics['total_return'].iloc[0] - metrics['buy_hold_return'].iloc[0]):.2%}")
print(f"\nSharpe Ratio: {metrics['sharpe_ratio'].iloc[0]:.2f}")
print(f"Max Drawdown: {metrics['max_drawdown'].iloc[0]:.2%}")
print(f"Win Rate: {metrics['win_rate'].iloc[0]:.2%}")
print(f"Expectancy: {metrics['expectancy'].iloc[0]:.4f}")
print(f"–ß–∏—Å–ª–æ —Å–¥–µ–ª–æ–∫: {int(metrics['num_trades'].iloc[0])}")
print("="*60)
```
