# –†–∞—Å—á–µ—Ç –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏—Ö –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π (Log Returns) –¥–ª—è –¥–∞–Ω–Ω—ã—Ö MOEX

–≠—Ç–æ—Ç –±–ª–æ–∫–Ω–æ—Ç —Å–æ–¥–µ—Ä–∂–∏—Ç —Å–∫—Ä–∏–ø—Ç –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏—Ö –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π –Ω–∞ –¥–∞–Ω–Ω—ã—Ö –ú–æ—Å–∫–æ–≤—Å–∫–æ–π –±–∏—Ä–∂–∏.

**–õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏** –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
```
log_return = ln(P_t / P_{t-1}) = ln(P_t) - ln(P_{t-1})
```

–ì–¥–µ:
- `P_t` - —Ü–µ–Ω–∞ –∑–∞–∫—Ä—ã—Ç–∏—è –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t
- `P_{t-1}` - —Ü–µ–Ω–∞ –∑–∞–∫—Ä—ã—Ç–∏—è –≤ –ø—Ä–µ–¥—ã–¥—É—â–∏–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏

**–ü—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–∞ log returns:**
- –°–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ—Å—Ç—å: –µ—Å–ª–∏ —Ü–µ–Ω–∞ –≤—ã—Ä–æ—Å–ª–∞ –Ω–∞ 10%, —Ç–æ log return = ln(1.1) ‚âà 0.0953
- –ï—Å–ª–∏ —Ü–µ–Ω–∞ —É–ø–∞–ª–∞ –Ω–∞ 10%, —Ç–æ log return = ln(0.9) ‚âà -0.1054
- –£–¥–æ–±—Å—Ç–≤–æ –¥–ª—è —Å—É–º–º–∏—Ä–æ–≤–∞–Ω–∏—è: —Å—É–º–º–∞ log returns –∑–∞ –ø–µ—Ä–∏–æ–¥ —Ä–∞–≤–Ω–∞ –æ–±—â–µ–π log return
- –°—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏–∏ (–≥–æ–º–æ—Å–∫–µ–¥–∞—Å—Ç–∏—á–Ω–æ—Å—Ç—å) –¥–ª—è –º–Ω–æ–≥–∏—Ö —Ñ–∏–Ω–∞–Ω—Å–æ–≤—ã—Ö –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤

````python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
````

````python
import os
import numpy as np
from pathlib import Path

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –ø—É—Ç–∏ –∫ –¥–∞–Ω–Ω—ã–º
# –ë–ª–æ–∫–Ω–æ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ ML/, –∞ –¥–∞–Ω–Ω—ã–µ –≤ ML/data/MOEX_DATA
# –í Jupyter —Ä–∞–±–æ—á–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –æ–±—ã—á–Ω–æ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–µ–π –±–ª–æ–∫–Ω–æ—Ç–∞
current_dir = Path.cwd()
DATA_DIR = current_dir / "data" / "MOEX_DATA"

# –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ –Ω–∞–π–¥–µ–Ω—ã, –ø—Ä–æ–±—É–µ–º –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –ø—É—Ç–∏
if not DATA_DIR.exists():
    # –ü—Ä–æ–±—É–µ–º –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –∫–æ—Ä–Ω—è –ø—Ä–æ–µ–∫—Ç–∞ (–µ—Å–ª–∏ –±–ª–æ–∫–Ω–æ—Ç –≤ ML/)
    alt_path = current_dir.parent / "ML" / "data" / "MOEX_DATA"
    if alt_path.exists():
        DATA_DIR = alt_path
    else:
        # –ü—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ –ø–∞–ø–∫—É data –≤ —Ç–µ–∫—É—â–µ–π –∏–ª–∏ —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–æ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
        for parent in [current_dir, current_dir.parent]:
            test_path = parent / "data" / "MOEX_DATA"
            if test_path.exists():
                DATA_DIR = test_path
                break

print(f"–¢–µ–∫—É—â–∞—è —Ä–∞–±–æ—á–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è: {current_dir}")
print(f"–ü—É—Ç—å –∫ –¥–∞–Ω–Ω—ã–º: {DATA_DIR}")
print(f"–°—É—â–µ—Å—Ç–≤—É–µ—Ç: {DATA_DIR.exists()}")

if not DATA_DIR.exists():
    print(f"\n‚ö†Ô∏è  –í–ù–ò–ú–ê–ù–ò–ï: –ü–∞–ø–∫–∞ {DATA_DIR} –Ω–µ –Ω–∞–π–¥–µ–Ω–∞!")
    print(f"–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –ø—É—Ç—å –∫ –¥–∞–Ω–Ω—ã–º.")
    print(f"\n–í–æ–∑–º–æ–∂–Ω—ã–µ –ø—É—Ç–∏:")
    print(f"  - {current_dir / 'data' / 'MOEX_DATA'}")
    print(f"  - {current_dir.parent / 'ML' / 'data' / 'MOEX_DATA'}")
else:
    print(f"‚úÖ –î–∞–Ω–Ω—ã–µ –Ω–∞–π–¥–µ–Ω—ã!")
    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Å–ø–∏—Å–æ–∫ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö —Ç–∏–∫–µ—Ä–æ–≤
    tickers = [d.name for d in DATA_DIR.iterdir() if d.is_dir() and not d.name.startswith('.')]
    print(f"–î–æ—Å—Ç—É–ø–Ω—ã–µ —Ç–∏–∫–µ—Ä—ã ({len(tickers)}): {', '.join(sorted(tickers)[:10])}{'...' if len(tickers) > 10 else ''}")
````

````python
def calculate_log_returns(prices):
    """
    –í—ã—á–∏—Å–ª—è–µ—Ç –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ (log returns).
    
    –§–æ—Ä–º—É–ª–∞: log(P_t / P_{t-1}) = log(P_t) - log(P_{t-1})
    
    Parameters:
    -----------
    prices : pd.Series
        –í–µ–∫—Ç–æ—Ä —Ü–µ–Ω (–æ–±—ã—á–Ω–æ —Ü–µ–Ω—ã –∑–∞–∫—Ä—ã—Ç–∏—è)
    
    Returns:
    --------
    pd.Series
        –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏. –ü–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –±—É–¥–µ—Ç NaN.
    """
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ Series, –µ—Å–ª–∏ —ç—Ç–æ –Ω–µ Series
    if not isinstance(prices, pd.Series):
        prices = pd.Series(prices)
    
    # –í—ã—á–∏—Å–ª—è–µ–º log returns
    log_prices = np.log(prices)
    log_returns = log_prices - log_prices.shift(1)
    
    return log_returns


def load_ticker_data(ticker, data_dir=DATA_DIR, timeframe='1D'):
    """
    –ó–∞–≥—Ä—É–∂–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ –ø–æ —Ç–∏–∫–µ—Ä—É –∏–∑ CSV —Ñ–∞–π–ª–∞.
    
    Parameters:
    -----------
    ticker : str
        –¢–∏–∫–µ—Ä –∞–∫—Ü–∏–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 'SBER', 'GAZP')
    data_dir : Path
        –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –¥–∞–Ω–Ω—ã–º–∏
    timeframe : str
        –¢–∞–π–º—Ñ—Ä–µ–π–º ('1D' –¥–ª—è –¥–Ω–µ–≤–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, '1H' –¥–ª—è —á–∞—Å–æ–≤—ã—Ö)
    
    Returns:
    --------
    pd.DataFrame
        DataFrame —Å –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
    """
    # –ü–æ–∏—Å–∫ —Ñ–∞–π–ª–∞
    ticker_dir = data_dir / ticker / timeframe
    if not ticker_dir.exists():
        raise FileNotFoundError(f"–ü–∞–ø–∫–∞ {ticker_dir} –Ω–µ –Ω–∞–π–¥–µ–Ω–∞")
    
    # –ü–æ–∏—Å–∫ CSV —Ñ–∞–π–ª–∞
    csv_files = list(ticker_dir.glob("*.csv"))
    if not csv_files:
        raise FileNotFoundError(f"CSV —Ñ–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ {ticker_dir}")
    
    csv_file = csv_files[0]
    print(f"–ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑: {csv_file}")
    
    # –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
    df = pd.read_csv(csv_file)
    
    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –¥–∞—Ç
    if 'end' in df.columns:
        df['date'] = pd.to_datetime(df['end'])
    elif 'begin' in df.columns:
        df['date'] = pd.to_datetime(df['begin'])
    
    # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –¥–∞—Ç–µ
    df = df.sort_values('date').reset_index(drop=True)
    
    return df

print("‚úÖ –§—É–Ω–∫—Ü–∏–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
````

````python
# –¢–µ—Å—Ç–∏—Ä—É–µ–º –Ω–∞ –¥–∞–Ω–Ω—ã—Ö SBER
ticker = 'SBER'
df_sber = load_ticker_data(ticker, timeframe='1D')

print(f"\n{'='*60}")
print(f"üìä –î–∞–Ω–Ω—ã–µ –¥–ª—è —Ç–∏–∫–µ—Ä–∞: {ticker}")
print(f"{'='*60}")
print(f"–ü–µ—Ä–∏–æ–¥: {df_sber['date'].min().date()} ‚Üí {df_sber['date'].max().date()}")
print(f"–í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {len(df_sber)}")
print(f"\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫:")
print(df_sber[['date', 'open', 'high', 'low', 'close', 'volume']].head())
print(f"\n–ü–æ—Å–ª–µ–¥–Ω–∏–µ 5 —Å—Ç—Ä–æ–∫:")
print(df_sber[['date', 'open', 'high', 'low', 'close', 'volume']].tail())
````

````python
# –í—ã—á–∏—Å–ª—è–µ–º log returns
df_sber['log_return'] = calculate_log_returns(df_sber['close'])

# –¢–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–∏–º –ø—Ä–æ—Å—Ç—ã–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
df_sber['simple_return'] = df_sber['close'].pct_change()

print(f"\n{'='*60}")
print("üìà –†–∞—Å—á–µ—Ç –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π")
print(f"{'='*60}")
print(f"\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ log returns:")
print(df_sber['log_return'].describe())
print(f"\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ simple returns:")
print(df_sber['simple_return'].describe())

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
print(f"\n{'='*60}")
print("–ü—Ä–∏–º–µ—Ä—ã –¥–∞–Ω–Ω—ã—Ö —Å –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—è–º–∏:")
print(f"{'='*60}")
display_cols = ['date', 'close', 'log_return', 'simple_return']
print(df_sber[display_cols].head(10))
print(f"\n...")
print(df_sber[display_cols].tail(10))
````

````python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –≥—Ä–∞—Ñ–∏–∫–∏ —Ü–µ–Ω –∏ log returns
import matplotlib.pyplot as plt
import seaborn as sns

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è (–ø—Ä–æ–±—É–µ–º —Ä–∞–∑–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã –¥–ª—è —Å–æ–≤–º–µ—Å—Ç–∏–º–æ—Å—Ç–∏)
try:
    plt.style.use('seaborn-v0_8-darkgrid')
except:
    try:
        plt.style.use('seaborn-darkgrid')
    except:
        try:
            plt.style.use('ggplot')
        except:
            pass  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –¥–µ—Ñ–æ–ª—Ç–Ω—ã–π —Å—Ç–∏–ª—å

sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 12)
plt.rcParams['font.size'] = 10

fig, axes = plt.subplots(3, 1, figsize=(14, 12))

# 1. –ì—Ä–∞—Ñ–∏–∫ —Ü–µ–Ω—ã –∑–∞–∫—Ä—ã—Ç–∏—è
ticker_name = ticker if 'ticker' in locals() else 'SBER'
axes[0].plot(df_sber['date'], df_sber['close'], linewidth=1.5, color='blue', label='–¶–µ–Ω–∞ –∑–∞–∫—Ä—ã—Ç–∏—è')
axes[0].set_title(f'–¶–µ–Ω–∞ –∑–∞–∫—Ä—ã—Ç–∏—è –∞–∫—Ü–∏–π {ticker_name}', fontsize=14, fontweight='bold')
axes[0].set_ylabel('–¶–µ–Ω–∞ (—Ä—É–±.)', fontsize=12)
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# 2. –ì—Ä–∞—Ñ–∏–∫ log returns
axes[1].plot(df_sber['date'], df_sber['log_return'], linewidth=0.8, color='green', alpha=0.7, label='Log Returns')
axes[1].axhline(y=0, color='red', linestyle='--', linewidth=1, alpha=0.5)
axes[1].set_title('–õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏–µ –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–∏ (Log Returns)', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Log Return', fontsize=12)
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

# 3. –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è log returns
axes[2].hist(df_sber['log_return'].dropna(), bins=50, color='purple', alpha=0.7, edgecolor='black')
axes[2].axvline(x=df_sber['log_return'].mean(), color='red', linestyle='--', linewidth=2, label=f'–°—Ä–µ–¥–Ω–µ–µ: {df_sber["log_return"].mean():.4f}')
axes[2].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∏—Ö –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç–µ–π', fontsize=14, fontweight='bold')
axes[2].set_xlabel('Log Return', fontsize=12)
axes[2].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n‚úÖ –ì—Ä–∞—Ñ–∏–∫–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω—ã")
print(f"–°—Ä–µ–¥–Ω—è—è log return: {df_sber['log_return'].mean():.6f}")
print(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {df_sber['log_return'].std():.6f}")
print(f"–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è log return: {df_sber['log_return'].min():.6f}")
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è log return: {df_sber['log_return'].max():.6f}")
````

````python
# –¢–µ—Å—Ç–∏—Ä—É–µ–º –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Ç–∏–∫–µ—Ä–∞—Ö
test_tickers = ['SBER', 'GAZP', 'LKOH', 'MTSS']

results = {}

for ticker in test_tickers:
    try:
        print(f"\n{'='*60}")
        print(f"–û–±—Ä–∞–±–æ—Ç–∫–∞ —Ç–∏–∫–µ—Ä–∞: {ticker}")
        print(f"{'='*60}")
        
        # –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
        df = load_ticker_data(ticker, timeframe='1D')
        
        # –í—ã—á–∏—Å–ª—è–µ–º log returns
        df['log_return'] = calculate_log_returns(df['close'])
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
        results[ticker] = {
            'data': df,
            'mean_log_return': df['log_return'].mean(),
            'std_log_return': df['log_return'].std(),
            'min_log_return': df['log_return'].min(),
            'max_log_return': df['log_return'].max(),
            'total_return': np.exp(df['log_return'].sum()) - 1,  # –û–±—â–∞—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—å –∑–∞ –ø–µ—Ä–∏–æ–¥
            'period_start': df['date'].min(),
            'period_end': df['date'].max(),
            'n_records': len(df)
        }
        
        print(f"‚úÖ {ticker}: {len(df)} –∑–∞–ø–∏—Å–µ–π, —Å—Ä–µ–¥–Ω—è—è log return: {results[ticker]['mean_log_return']:.6f}")
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ {ticker}: {e}")

print(f"\n{'='*60}")
print("üìä –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤")
print(f"{'='*60}")

# –°–æ–∑–¥–∞–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
summary_data = []
for ticker, stats in results.items():
    summary_data.append({
        '–¢–∏–∫–µ—Ä': ticker,
        '–ó–∞–ø–∏—Å–µ–π': stats['n_records'],
        '–ü–µ—Ä–∏–æ–¥ –Ω–∞—á–∞–ª–æ': stats['period_start'].date(),
        '–ü–µ—Ä–∏–æ–¥ –∫–æ–Ω–µ—Ü': stats['period_end'].date(),
        '–°—Ä–µ–¥–Ω—è—è log return': f"{stats['mean_log_return']:.6f}",
        '–°—Ç. –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ': f"{stats['std_log_return']:.6f}",
        '–ú–∏–Ω. log return': f"{stats['min_log_return']:.6f}",
        '–ú–∞–∫—Å. log return': f"{stats['max_log_return']:.6f}",
        '–û–±—â–∞—è –¥–æ—Ö–æ–¥–Ω–æ—Å—Ç—å %': f"{(stats['total_return'] * 100):.2f}%"
    })

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))
````

````python
# –°—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è log returns –¥–ª—è –≤—Å–µ—Ö —Ç–∏–∫–µ—Ä–æ–≤
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx, (ticker, stats) in enumerate(results.items()):
    if idx >= 4:
        break
    
    df = stats['data']
    
    # –ì—Ä–∞—Ñ–∏–∫ log returns –≤–æ –≤—Ä–µ–º–µ–Ω–∏
    axes[idx].plot(df['date'], df['log_return'], linewidth=0.8, alpha=0.7, label=f'{ticker} Log Returns')
    axes[idx].axhline(y=0, color='red', linestyle='--', linewidth=1, alpha=0.5)
    axes[idx].set_title(f'{ticker} - Log Returns', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Log Return', fontsize=10)
    axes[idx].legend()
    axes[idx].grid(True, alpha=0.3)
    axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π
fig, ax = plt.subplots(figsize=(14, 8))

for ticker, stats in results.items():
    df = stats['data']
    ax.hist(df['log_return'].dropna(), bins=50, alpha=0.5, label=f'{ticker} (œÉ={stats["std_log_return"]:.4f})', density=True)

ax.set_title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π log returns –¥–ª—è —Ä–∞–∑–Ω—ã—Ö —Ç–∏–∫–µ—Ä–æ–≤', fontsize=14, fontweight='bold')
ax.set_xlabel('Log Return', fontsize=12)
ax.set_ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', fontsize=12)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("‚úÖ –°—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω—ã")
````

````python
# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∏ –ø—Ä–æ–≤–µ—Ä–∫–∏

print(f"\n{'='*60}")
print("üîç –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∏ –ø—Ä–æ–≤–µ—Ä–∫–∏")
print(f"{'='*60}")

# –ü—Ä–æ–±—É–µ–º –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å scipy –¥–ª—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö —Ç–µ—Å—Ç–æ–≤
try:
    from scipy import stats as scipy_stats
    HAS_SCIPY = True
except ImportError:
    HAS_SCIPY = False
    print("‚ö†Ô∏è  scipy –Ω–µ —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω. –ù–µ–∫–æ—Ç–æ—Ä—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ç–µ—Å—Ç—ã –±—É–¥—É—Ç –ø—Ä–æ–ø—É—â–µ–Ω—ã.")
    print("   –î–ª—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏: pip install scipy")

for ticker, stats in results.items():
    df = stats['data']
    log_returns = df['log_return'].dropna()
    
    print(f"\n{ticker}:")
    print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π: {len(log_returns)}")
    print(f"  –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è: {df['log_return'].isna().sum()}")
    print(f"  –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∞—Å–∏–º–º–µ—Ç—Ä–∏–∏ (skewness): {log_returns.skew():.4f}")
    print(f"  –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —ç–∫—Å—Ü–µ—Å—Å–∞ (kurtosis): {log_returns.kurtosis():.4f}")
    print(f"  –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {log_returns.mean():.6f}")
    print(f"  –ú–µ–¥–∏–∞–Ω–∞: {log_returns.median():.6f}")
    print(f"  –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {log_returns.std():.6f}")
    print(f"  –ú–∏–Ω–∏–º—É–º: {log_returns.min():.6f}")
    print(f"  –ú–∞–∫—Å–∏–º—É–º: {log_returns.max():.6f}")
    print(f"  25% –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å: {log_returns.quantile(0.25):.6f}")
    print(f"  75% –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å: {log_returns.quantile(0.75):.6f}")
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å (—Ç–æ–ª—å–∫–æ –µ—Å–ª–∏ scipy –¥–æ—Å—Ç—É–ø–µ–Ω)
    if HAS_SCIPY:
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ (–¥–ª—è –Ω–µ–±–æ–ª—å—à–∏—Ö –≤—ã–±–æ—Ä–æ–∫ –¥–æ 5000 –Ω–∞–±–ª—é–¥–µ–Ω–∏–π)
        if len(log_returns) <= 5000:
            try:
                shapiro_stat, shapiro_p = scipy_stats.shapiro(log_returns)
                print(f"  –¢–µ—Å—Ç –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å:")
                print(f"    statistic={shapiro_stat:.4f}, p-value={shapiro_p:.6f}")
                if shapiro_p < 0.05:
                    print(f"    ‚ö†Ô∏è  –û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (p < 0.05)")
                else:
                    print(f"    ‚úÖ –î–∞–Ω–Ω—ã–µ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã (p >= 0.05)")
            except Exception as e:
                print(f"    ‚ö†Ô∏è  –û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ —Ç–µ—Å—Ç–∞ –®–∞–ø–∏—Ä–æ-–£–∏–ª–∫–∞: {e}")

print(f"\n{'='*60}")
print("‚úÖ –ê–Ω–∞–ª–∏–∑ –∑–∞–≤–µ—Ä—à–µ–Ω")
print(f"{'='*60}")
print("\nüìù –†–µ–∑—é–º–µ:")
print("  - Log returns —É—Å–ø–µ—à–Ω–æ –≤—ã—á–∏—Å–ª–µ–Ω—ã –¥–ª—è –≤—Å–µ—Ö —Ç–∏–∫–µ—Ä–æ–≤")
print("  - –ü—Ä–æ–≤–µ–¥–µ–Ω–∞ –±–∞–∑–æ–≤–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞")
print("  - –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π 'results'")
print("  - –î–ª—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö —Ç–µ—Å—Ç–æ–≤ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ scipy: pip install scipy")
````
